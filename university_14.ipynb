```python
def Mean(x, y):
    AMean = (x + y) / 2
    GMean = (x * y) ** 0.5
    return AMean, GMean
```

```python
import math

def Circle(R):
    S = math.pi * R ** 2
    L = 2 * math.pi * R
    return S, L
```

```python
def TrianglePS(a):
    P = 3 * a
    S = (3 ** 0.5 / 4) * a ** 2
    return P, S
```

```python
import math

def RingS(R1, R2):
    S = math.pi * (R1 ** 2 - R2 ** 2)
    return S
```

```python
def RectPS(x1, y1, x2, y2):
    a = abs(x2 - x1)
    b = abs(y2 - y1)
    P = 2 * (a + b)
    S = a * b
    return P, S
```

```python
def TriangleP(a, h):
    b = ((a / 2) ** 2 + h ** 2) ** 0.5
    P = a + 2 * b
    return P
```

```python
def InvertDigits(K):
    return int(str(K)[::-1])
```

```python
def SumRange(A, B):
    if A > B:
        return 0
    else:
        return sum(range(A, B + 1))
```

```python
def TypeTrl(x1, y1, x2, y2, x3, y3):
    def side(xa, ya, xb, yb):
        return ((xa - xb) ** 2 + (ya - yb) ** 2) ** 0.5

    a = side(x2, y2, x3, y3)
    b = side(x1, y1, x3, y3)
    c = side(x1, y1, x2, y2)

    # Проверка на равенство сторон (равносторонний, равнобедренный)
    if abs(a - b) < 1e-6 and abs(b - c) < 1e-6:
        return "Равносторонний"
    elif abs(a - b) < 1e-6 or abs(b - c) < 1e-6 or abs(a - c) < 1e-6:
        t = sorted([a, b, c])
        if abs(t[0] ** 2 + t[1] ** 2 - t[2] ** 2) < 1e-6:
            return "Равнобедренный прямоугольный"
        return "Равнобедренный"
    else:
        t = sorted([a, b, c])
        if abs(t[0] ** 2 + t[1] ** 2 - t[2] ** 2) < 1e-6:
            return "Прямоугольный"
        return "Обычный"
```

```python
def Quarter(x, y):
    if x > 0 and y > 0:
        return 1
    elif x < 0 and y > 0:
        return 2
    elif x < 0 and y < 0:
        return 3
    else:
        return 4
```

```python
def DigitCountSum(K):
    digits = [int(i) for i in str(K)]
    count = len(digits)
    total = sum(digits)
    return count, total
```

```python
def Calc(A, B, Op):
    if Op == 1:
        return A - B
    elif Op == 2:
        return A * B
    elif Op == 3:
        return A / B
    else:
        return A + B
```

```python
def LuckyNum(n):
    s = str(n)
    return int(s[0]) + int(s[1]) == int(s[2]) + int(s[3])

def AllLuckyNums():
    lucky = []
    for n in range(1000, 10000):
        if LuckyNum(n):
            lucky.append(n)
    return lucky
```

```python
import math

def DegToRad(D):
    return D * math.pi / 180
```

```python
def IsLeapYear(Y):
    return (Y % 4 == 0 and Y % 100 != 0) or (Y % 400 == 0)
```
