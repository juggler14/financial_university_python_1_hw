```python
#1
def Fact(N):
    if N == 0 or N == 1:
        return 1
    return N * Fact(N - 1)

def Fact2(N):
    if N <= 1:
        return 1
    return N * Fact2(N - 2)

n = int(input())
print(Fact(n), Fact2(n))
```

```python
#2
def PowerN(x, n):
    if n == 0:
        return 1
    elif n > 0:
        return x * PowerN(x, n - 1)
    else:
        return 1 / PowerN(x, -n)

x, n = float(input()), int(input()) 
print(PowerN(x, n))
```

```python
#3
def SqrtK(x, k, n):
    if n == 0:
        return 1
    y_prev = SqrtK(x, k, n - 1)
    return y_prev - (y_prev - x / (y_prev ** (k - 1))) / k

x, k, n = float(input()), int(input()), int(input())
print(SqrtK(x, k, n))
```

```python
#4
call_count = 0

# Классическая рекурсивная формула Фибоначчи F(n) = F(n-1) + F(n-2)
def FibRec(N):
    global call_count
    call_count += 1
    
    if N == 1 or N == 2:
        return 1
    return FibRec(N - 2) + FibRec(N - 1)

# Ввод 5 чисел и вывод каждого с количеством рекурсивных вызовов
numbers = list(map(int, input().split()))

for num in numbers[:5]:
    call_count = 0
    fib_value = FibRec(num)
    print(fib_value, call_count)
```

```python
#5
c_call_count = 0

# Рекурсивное вычисление сочетаний по формуле: C(m,n) = C(m,n-1) + C(m-1,n-1)
def C(m, n):
    global c_call_count
    c_call_count += 1
    
    if m == 0 or m == n:
        return 1
    if 0 < m < n:
        return C(m, n - 1) + C(m - 1, n - 1)
    return 0

N = int(input())
M_values = list(map(int, input().split()))

for m in M_values[:5]:
    c_call_count = 0
    comb_value = C(m, N)
    print(comb_value, c_call_count)
```

```python
#6
# Алгоритм Евклида: НОД(A,B) = НОД(B mod A, A)
def NOD(A, B):
    if A == 0:
        return B
    return NOD(B % A, A)

A, B, C, D = map(int, input().split())
print(NOD(A, B), NOD(A, C), NOD(A, D))
```

```python
#7
# Поиск минимума путем сравнения последнего элемента с минимумом остальной части
def MinRec(A, N):
    if N == 1:
        return A[0]
    min_rest = MinRec(A, N - 1)
    return A[N - 1] if A[N - 1] < min_rest else min_rest

arrays = []
for _ in range(3):
    arr = list(map(int, input().split()))
    arrays.append(arr)

results = [MinRec(arr, len(arr)) for arr in arrays]
print(*results)
```

```python
#8
# Рекурсивный подсчет цифр в строке
def Digits(S):
    if not S:
        return 0
    return (1 if S[0].isdigit() else 0) + Digits(S[1:])

S = input()
print(Digits(S))
```

```python
#9
# Проверка всей строки и части строки на палиндром
def Simm(S):
    if len(S) <= 1:
        return True
    if S[0] != S[-1]:
        return False
    return Simm(S[1:-1])

def Simm_partial(S, i, j):
    if i >= j:
        return True
    if S[i] != S[j]:
        return False
    return Simm_partial(S, i + 1, j - 1)

S = input()
i, j = map(int, input().split())

print("yes" if Simm(S) else "no")
print("yes" if Simm_partial(S, i, j) else "no")
```

```python
#10
# Побитовое сложение двоичных строк с переносом
def add_binary(a, b):
    max_len = max(len(a), len(b))
    a = a.zfill(max_len)
    b = b.zfill(max_len)
    
    result = ''
    carry = 0
    
    for i in range(max_len - 1, -1, -1):
        bit_sum = carry + (a[i] == '1') + (b[i] == '1')
        result = ('1' if bit_sum % 2 == 1 else '0') + result
        carry = bit_sum // 2
    
    if carry:
        result = '1' + result
    
    return result.lstrip('0') or '0'

positive, negative = input().split()
negative = negative.lstrip('-')
print(add_binary(positive, negative))
```

```python
#11
# Метод бисекции для нахождения корня f(x)=0 на отрезке [a,b]
def f(x):
    return x**3 - 2*x - 5

def Root(f, a, b, epsilon):
    mid = (a + b) / 2
    if abs(b - a) < epsilon or abs(f(mid)) < epsilon:
        return mid
    
    if f(a) * f(mid) < 0:
        return Root(f, a, mid, epsilon)
    else:
        return Root(f, mid, b, epsilon)

a, b, epsilon = map(float, input().split())
print(Root(f, a, b, epsilon))
```

```python
#12
# Рекурсивный переворот числа без использования строк
def reverse_number(n):
    def helper(n, rev):
        if n == 0:
            return rev
        return helper(n // 10, rev * 10 + n % 10)
    
    return helper(n, 0)

n = int(input())
print(reverse_number(n))
```

```python
#14

# Рекурсивное вычисление суммы цифр
def sum_digits(N):
    if N == 0:
        return 0
    return N % 10 + sum_digits(N // 10)

N = int(input())
print(sum_digits(N))
```

```python
#15
# Рекурсивная проверка слова на палиндром
def is_palindrome(word):
    if len(word) <= 1:
        return "yes"
    if word[0] != word[-1]:
        return "no"
    return is_palindrome(word[1:-1])

word = input()
print(is_palindrome(word))
```
