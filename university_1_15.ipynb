```python
#1
def remove_row_col(matrix, k, l):
    if not matrix or k >= len(matrix) or l >= len(matrix[0]):
        return matrix
    return [[matrix[i][j] for j in range(len(matrix[0])) if j != l] 
            for i in range(len(matrix)) if i != k]
```

```python
#2
def split_list(a):
    # Разделяет на четные и нечетные
    even = [x for x in a if x % 2 == 0]
    odd = [x for x in a if x % 2 != 0]
    return even, odd

# Пример использования:
# even, odd = split_list([1, 2, 3, 4, 5, 6])
# print(f"Четные: {even}, Нечетные: {odd}")
```

```python
#3
def swap_row(matrix, k1, k2):
    # Меняет местами строки K1 и K2 в матрице
    if not matrix or k1 >= len(matrix) or k2 >= len(matrix):
        return matrix
    
    matrix[k1], matrix[k2] = matrix[k2], matrix[k1]
    return matrix

# matrix = [[1, 2], [3, 4], [5, 6]]
# swap_row(matrix, 0, 2)
# print(matrix)  # [[5, 6], [3, 4], [1, 2]]
```

```python
#5
def mtr_prod(a, b):

    if len(a[0]) != len(b):
        return 0
    
    m, n, p = len(a), len(b), len(b[0])
    result = [[0 for _ in range(p)] for _ in range(m)]
    
    for i in range(m):
        for j in range(p):
            for k in range(n):
                result[i][j] += a[i][k] * b[k][j]
    
    return result

"""
a = [[1, 2], [3, 4]]
b = [[5, 6], [7, 8]]
result = mtr_prod(a, b)
print(result)  # [[19, 22], [43, 50]]
"""
```

```python
#6
def sum_mtr(a, b):

    if len(a) != len(b) or len(a[0]) != len(b[0]):
        raise ValueError("пу пу пу")
    
    m, n = len(a), len(a[0])
    return [[a[i][j] + b[i][j] for j in range(n)] for i in range(m)]

# a = [[1, 2], [3, 4]]
# b = [[5, 6], [7, 8]]
# result = sum_mtr(a, b)
# print(result)  # [[6, 8], [10, 12]]
```

```python
#7
def diagonal_sum(matrix, diagonal_type='main'):
    n = len(matrix)
    return sum(matrix[i][i if diagonal_type == 'main' else n-1-i] for i in range(n))
```

```python
#8
def remove_rows(matrix, k1, k2):
    if not matrix: return matrix
    k2 = min(k2, len(matrix)-1)
    return [row for i, row in enumerate(matrix) if i < k1 or i > k2]
```

```python
#9
def invert_str(s, k, n):
    return "" if k >= len(s) else s[k:k+n][::-1]
```

```python
#10
def swap_col(matrix, k1, k2):
    """Меняет местами столбцы K1 и K2 в матрице"""
    if not matrix or k1 >= len(matrix[0]) or k2 >= len(matrix[0]):
        return matrix
    
    for i in range(len(matrix)):
        matrix[i][k1], matrix[i][k2] = matrix[i][k2], matrix[i][k1]
    
    return matrix

# matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
# swap_col(matrix, 0, 2)
# print(matrix)  # [[3, 2, 1], [6, 5, 4], [9, 8, 7]]
```

```python
#11
from itertools import groupby

def compress_str(s):
    result = []
    for char, group in groupby(s):
        count = len(list(group))
        result.append(f"{char}{{{count}}}" if count > 4 else char * count)
    return "".join(result)
```

```python
#12
def is_ident(s):
    return bool(s) and not s[0].isdigit() and all(c.isalnum() or c == '_' for c in s)
```

```python
#13
def chessboard(m, n):
    return [[(i + j) % 2 for j in range(n)] for i in range(m)]
```

```python
#14
def bell(arr):
    sorted_arr = sorted(arr)
    result = []
    for a, b in zip(sorted_arr, reversed(sorted_arr)):
        result.extend([a, b])
    return result[:len(sorted_arr)]
```

```python
#15
def remove_cols(matrix, k1, k2):
    if not matrix: return matrix
    k2 = min(k2, len(matrix[0])-1)
    return [[row[j] for j in range(len(row)) if j < k1 or j > k2] for row in matrix]
```
