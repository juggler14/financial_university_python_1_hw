```python
#1
def Fact(N):
    if N == 0 or N == 1:
        return 1
    return N * Fact(N - 1)

def Fact2(N):
    if N == 1 or N == 2:
        return N
    return N * Fact2(N - 2)

n = int(input())
print(f"fact1 - {Fact(n)} \nfact2 - {Fact2(n)}")
```

```python
#2
def PowerN(x, n):
    if n == 0:
        return 1
    elif n > 0:
        return x * PowerN(x, n - 1)
    else:
        return 1 / PowerN(x, -n)

x = int(input())
n = int(input())
print(f"PowerN {Fact(x, n)}")
```

```python
#3
def SqrtK(x, k, n):
    if n == 0:
        return 1
    y = SqrtK(x, k, n - 1)
    return y - (y - x / (y ** (k - 1))) / k

# Начинаем с начального приближения y(0) = 1

# На каждом шаге улучшаем приближение по формуле

# Чем больше n (глубина рекурсии), тем точнее результат

x, k, n = map(int, input().split())
print(SqrtK(x, k, n))
```

```python
#5
call_count = 0

def Fib(N):
    global call_count
    call_count += 1
    if N == 1 or N == 2:
        return 1
    return Fib(N - 2) + Fib(N - 1)

# Пример(5)
numbers = [5, 7, 10, 12, 15]
for n in numbers:
    call_count = 0
    result = Fib(n)
    print(f"F({n}) = {result}, вызовов: {call_count}")
```

```python
#6
def NOD(A, B):
    if A == 0:
        return B
    return NOD(B % A, A)

# Пример
A, B, C, D = 36, 48, 60, 72
print(f"NOD({A},{B}) = {NOD(A, B)}")
print(f"NOD({A},{C}) = {NOD(A, C)}")
print(f"NOD({A},{D}) = {NOD(A, D)}")
```

```python
#7
def MinRec(A, N):
    if N == 1:
        return A[0]
    return min(A[N - 1], MinRec(A, N - 1))

# Пример
arr1 = [3, 1, 4, 1, 5]
arr2 = [9, 2, 6, 1, 8]
arr3 = [5, 2, 9, 3, 7]
print(f"Min1: {MinRec(arr1, len(arr1))}")
print(f"Min2: {MinRec(arr2, len(arr2))}")
print(f"Min3: {MinRec(arr3, len(arr3))}")
```

```python
#8
def Digits(S):
    if S == "":
        return 0
    return (1 if S[0].isdigit() else 0) + Digits(S[1:])

# Пример
s = "abc123def45"
print(Digits(s))
```

```python
#9
def Simm(S):
    if len(S) <= 1:
        return True
    if S[0] != S[-1]:
        return False
    return Simm(S[1:-1])

# Пример
test_str = "radar"
print(f"'{test_str}' - {Simm(test_str)}")
```

```python
#10
def binary_add(a, b):
    # Преобразуем в десятичные, складываем и обратно в двоичное
    dec_a = int(a, 2) if a[0] != '-' else -int(a[1:], 2)
    dec_b = int(b, 2) if b[0] != '-' else -int(b[1:], 2)
    result = dec_a + dec_b
    return bin(result)[2:] if result >= 0 else '-' + bin(-result)[2:]

# Пример
pos_bin = "1010"  # 10
neg_bin = "-101"  # -5
print(binary_add(pos_bin, neg_bin))  # 101 (5)
```

```python
#11
def Root(f, a, b, eps):
    mid = (a + b) / 2
    if abs(b - a) < eps or abs(f(mid)) < eps:
        return mid
    if f(a) * f(mid) < 0:
        return Root(f, a, mid, eps)
    else:
        return Root(f, mid, b, eps)

# Пример для f(x) = x^2 - 2
def f(x):
    return x**2 - 2

print((f, 0, 2, 0.0001):.6f)
```

```python
#12
def ReverseDigits(n):
    if n < 10:
        return n
    last_digit = n % 10
    remaining = n // 10
    reversed_remaining = ReverseDigits(remaining)
    # Подсчитываем разрядность reversed_remaining
    digits = len(str(reversed_remaining))
    return last_digit * (10 ** digits) + reversed_remaining

# Пример
num = 12345
print(ReverseDigits(num))
```

```python
#13
def Simm_sub(S, i, j):
    if i >= j:
        return True
    if S[i] != S[j]:
        return False
    return Simm_sub(S, i + 1, j - 1)

# Пример
s = "racecar"
print({Simm_sub(s, 1, 5)})
```

```python
#14
def SumDigits(N):
    if N == 0:
        return 0
    return N % 10 + SumDigits(N // 10)

# Пример
num = 12345
print(f"Сумма - {SumDigits(num)}")
```

```python
#15
def is_palindrome(word):
    if len(word) <= 1:
        return "yes"
    if word[0] != word[-1]:
        return "no"
    return is_palindrome(word[1:-1])

# Пример
word = "racecar"
print(is_palindrome(word))
```
