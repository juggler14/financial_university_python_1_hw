```python
# 1
import math

a = []
n = int(input())
for i in range(n):
    element = int(input())
    a.append(element)
k, l = map(int, input().split())
if not (1 <= k <= l <= n):
    print('error')
masa2 = a[k - 1:l]
sumel = sum(masa2)
countel = len(masa2)
srarif = sumel / countel
print(srarif)
```

```python
#4
a = [0, 1, 2, 3, 4, 5]
chet = []
nechet = []
for i in range(0, len(a)):
    if a[i] % 2 == 0:
        chet.append(a[i])
    else:
        nechet.append(a[i])
nechet.reverse()
maxar = chet + nechet
print(maxar)
```

```python
# 7
a = [3, 1, 2, 3, 4, 7]
maxel = max(a)
minel = min(a)
r = a.index(maxel)
l = a.index(minel)
if r - l > 0:
    masa = a[l:r + 1]
else:
    masa = a[r:l + 1]
print(masa)
suma = sum(masa)
print(suma)
```

```python
# 8
a = [3, 1, 2, 3, 4, 7]
maxel = max(a)
minel = min(a)
r = a.index(maxel)
l = a.index(minel)
start = min(r, l)
finish = max(l, r)
for i in range(start + 1, finish):
    a[i] = 0
print(a)
```
