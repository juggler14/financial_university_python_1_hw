```python
#(a)
def parts_from_sequence(s):
    
    digits = set(ch for ch in s if ch.isdigit())  
    znaks = set(ch for ch in s if ch in '+-*/')     
    letters_AF = set(ch for ch in s if ch.upper() in set('ABCDEF') )
    letters_XZ = set(ch for ch in s if ch.upper() in set('XYZ') )
    
    return digits, znaks, letters_AF, letters_XZ
```

```python
# (b)
def missing_products(given_codes, all_codes=None):
    
    if all_codes is None:
        all_codes = set(range(1, 21))
    else:
        all_codes = set(all_codes)
        
    present = set(given_codes)
    missing = all_codes - present
    return missing
```
