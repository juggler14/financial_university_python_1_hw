### Блок 4.1

### Блок 4.13

```python
n = int(input())
maxi = 1
for i in range(1, n + 1):
    if i ** 2 <= n:
        maxi = i ** 2
print(maxi)
```

### Блок 4.14

```python
n = int(input())
maxk = 1
for k in range(1, n + 1):
    if 3 ** k < n:
        maxk = k
print(maxk)
```

### Блок 4.15

```python
n = int(input())
a, b = map(float, input().split())
lined = (b - a) / n
print(lined)
for i in range(0, int(b - a)):
    print(a + lined * i)
```
